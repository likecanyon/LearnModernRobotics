# 第二章 位形空间

### 自由度定义

机器人的自由度（degree of freedom, dof）数就是表示其位形所需的最小实数值坐标数。

> 注意是实数值，不是离散值，比如硬币正反面{正、反}就不是一个自由度

### 位形空间

机器人的**位形**是指机器人上每个点位的全体集合，表示机器人位形所需的最少广义坐标数$n$就是机器人的自由度（dof），包含所有可能的机器人位形的$n$维空间成为位形空间（C-空间，Configuration Space）。机器人的位形通常用C-空间中的一个特征点来表示。


## 2.1 刚体的自由度

$$
系统自由度 = (点的自由度之和) - (独立约束数) = (变量数) - (独立约束方程数)
$$

- 三维空间中运动的刚体，称之为**空间刚体**
- 二维平面中运动的刚体，称之为**平面刚体**

## 2.2 机器人的自由度

### Grubler公式

对于一个具有$N$个构件（含基座）的机构，令$J$为关节数，$m$为刚体的自由度数（对于平面机构，$m=3$；对于空间机构，$m=6$），$f_i$为关节$i$对应的自由度数，$c_i$为其对应的约束数，对于所有的$i$，式中满足$f_i+c_i=m$。则用于计算机器人自由度的Gruber公式可以写成：

$$
\text{dof}=\underbrace {m(N - 1)}_{\text{刚体自由度数}} - \underbrace {\sum_{i=1}^{J}c_i}_{\text{关节约束数}} \

=m(N-1)-\sum_{i=1}^{J}(m-f_i) \
=m(N-1-J)+\sum_{i=1}^{J}(f_i) 
$$

> 该公式只有在所有关节约束都是独立的情况下才能成立。否则，该公式只能用于判断自由度数的下限值。


**冗余约束与奇异**：冗余约束不具备独立性，某些构型下存在奇异，Grubler公式只能提供一个自由度下限值。

- Delta机器人：15个自由度，3个为动平台空间移动自由度（有用的实际自由度），还有12个局部自由度是12根杆绕自身长轴旋转
- Stewart-Gough平台：6个自由度，并联结构，作为汽车及飞机模拟器非常受欢迎


## 2.3 位形空间拓扑表达

**拓扑等效**：如果两个表面能够连续从一种形状变化到另一种形状（不能采用切割和黏结等方式来实现），则称之为拓扑等效。

不同拓扑的一维空间包括：圆、直线、线段。圆在数学上写成$S^1$，直线写成$\mathbb{E}^1$，线段$[a,b]\in \mathbb{R^1}$


### 位形空间的表达

$n$个坐标或者参数表示$n$维空间的方法称为空间的显示参数化（explicit parametrization），显示表达会存在**奇异（Singularity）**情况

空间隐式表达（implicit representation），使用高维空间的坐标，缺点是参数比自由度数多，优点在于无奇异。另一个优点：对于闭链结构，无论构建显示参数化方程，还是绘出图册都是非常困难，**但找到其隐式表达相对比较容易**。满足**闭环方程（loop-closure equation）**

**旋转矩阵**：9个参数，6个约束方程，表示空间3个姿态的旋转

C-空间可以视为嵌入$n$维空间中的$n-k$维曲面（$k$为完整约束，独立约束方程）

## 位形与速度约束

**Pfaffian约束**：形式如$A(\theta)\dot{\theta}$的速度约束方程

形如$g(\theta)=0$的完整约束也称为**可积约束（integrable constraint）**

不可积的Pfaffian约束称作非完整约束（nonholonomic constraint）。可以减少系统速度的维数，但不会减少可达C-空间维数

## 2.5 任务空间与工作空间

- **任务空间**：机器人的任务能够自然表达的空间，定义任务空间决定因素在任务本身，与机器人无关
- **工作空间**：机器人末端执行器所能到达位形的指标，主要取决于机器人结构
